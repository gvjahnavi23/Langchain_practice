```python
from langchain_huggingface import HuggingFaceEndpoint, ChatHuggingFace
from google.colab import userdata
HUGGINGFACE_API_TOKEN = userdata.get('HF_TOKEN')

llm = HuggingFaceEndpoint(
    repo_id="google/gemma-4-31B-it",
    task="text-generation",
    huggingfacehub_api_token = HUGGINGFACE_API_TOKEN
)

model = ChatHuggingFace(
    llm = llm
)
```

```python
from langchain_core.prompts import PromptTemplate

template = PromptTemplate(
    input_variables=["input"],
    template="Explain {input}"
)
```

```python
model.invoke(template.invoke('llm'))
```

**String output parser**

```python
from langchain_core.output_parsers import StrOutputParser

string_parser = StrOutputParser()

chain = template | model | string_parser

chain.invoke('llm')
```

**JSON output parser**

```python
#classic json output parser

from langchain_core.output_parsers import JsonOutputParser
from langchain_core.prompts import PromptTemplate

json_parser1 = JsonOutputParser()
template = PromptTemplate(
    input_variables=["input"],
    template="Explain {input}\n{format_instructions}",
    partial_variables={
        "format_instructions": json_parser1.get_format_instructions()
    }
)


chain2 = template | model | json_parser1

chain2.invoke('llm')
```

**Enforced json object**

```python
from pydantic import BaseModel, Field

class Summary(BaseModel):
  topic: str = Field(description="The main topic of the input")
  summary: str = Field(description="A short summary of the input")
  key_point1: str = Field(description="A key point of the input")
  key_point2: str = Field(description="A key point of the input")
```

```python
from langchain_core.prompts import PromptTemplate
from langchain_core.output_parsers import JsonOutputParser

json_parser2 = JsonOutputParser(pydantic_object=Summary)

template = PromptTemplate(
    input_variables=["input"],
    template="Explain {input}\n{format_instructions}",
    partial_variables={
        "format_instructions": json_parser2.get_format_instructions()
    }
)

chain2 = template | model | json_parser2

chain2.invoke('llm')
```

```python
from pydantic import BaseModel, Field

class Summary2(BaseModel):
  topic: str = Field(description="The main topic of the input")
  summary: str = Field(description="A short summary of the input")
  key_point1: str = Field(description="A key point of the input")
  key_point2: str = Field(description="A key point of the input")
  key_point3: str = Field(description="A key point of the input")
  key_point4: str = Field(description="A key point of the input")
  key_point5: str = Field(description="A key point of the input")
```

```python
from langchain_core.output_parsers import PydanticOutputParser
from langchain_core.prompts import PromptTemplate

pydantic_parser = PydanticOutputParser(pydantic_object=Summary2)
template = PromptTemplate(
    input_variables=["input"],
    template="Explain {input}\n{format_instructions}",
    partial_variables={
        "format_instructions": pydantic_parser.get_format_instructions()
    }
)

chain3 = template | model | pydantic_parser

chain3.invoke('llm')
```
