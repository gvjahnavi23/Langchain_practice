```python
import os
from langchain_google_genai import GoogleGenerativeAI
from google.colab import userdata
gemini_api = userdata.get('gemini_api')

llm = GoogleGenerativeAI(model="gemini-2.5-flash", api_key = gemini_api, temperature = 0.2, max_retries = 3)
# Invoke the model directly
response = llm.invoke("what is the capital of India?")

# Access the text output comfortably
print(response)
```

```python
for chunk in llm.stream("what is the capital of India?"):
    print(chunk, end=" ", flush=True)
```

```python
from langchain_google_genai import ChatGoogleGenerativeAI

chat_model1 = ChatGoogleGenerativeAI(
    model="gemini-2.5-flash",
    api_key = gemini_api,
    temperature = 0.2,
    max_retries = 3
)

result = chat_model1.invoke("what is the capital of India?")
result
```

```python
from langchain_huggingface import HuggingFaceEndpoint
from google.colab import userdata

huggingface_api = userdata.get('HF_TOKEN')

normal_model = HuggingFaceEndpoint(
    repo_id="mistralai/Mistral-7B-v0.1",
    task="text-generation",
    huggingfacehub_api_token=huggingface_api
)

result = normal_model.invoke("what is the capital of India?")
print(result)
```

```python
from langchain_huggingface import ChatHuggingFace
from langchain_huggingface import HuggingFaceEndpoint


llm2 = HuggingFaceEndpoint(
    repo_id="Qwen/Qwen2.5-7B-Instruct",
    huggingfacehub_api_token=huggingface_api
)

chat_model2 = ChatHuggingFace(
    llm = llm2
)

chat_model2.invoke("what is the capital of India?")
```

```python
from langchain_core.prompts import PromptTemplate

template = PromptTemplate(
    template = """what is the capital of {country} and also give information about the capital city?""",
    input_variables = ["country"],
    validate_template = True
)
```

```python
country = input('Enter the name of the country you want to know the capital for')

prompt = template.invoke({
    'country' : country,
}
)

chat_model1.invoke(prompt)
```
