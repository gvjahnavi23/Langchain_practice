**Structured output**

```python
from typing import TypedDict, Optional, List

class Review(TypedDict):
  review: str
  sentiment: str
  key_points: Optional[List[str]]
```

```python
from langchain_google_genai import ChatGoogleGenerativeAI

from google.colab import userdata
gemini_api = userdata.get('gemini_api')

llm = ChatGoogleGenerativeAI(model="gemini-2.5-flash", api_key = gemini_api, temperature = 0.4, max_retries = 3)

print("Gemini model initialized successfully!")
```

```python
structured_llm = llm.with_structured_output(Review)
```

```python
structured_llm.invoke(
    "I’ve been using the AeroGlide Cordless Vacuum for about two weeks now. Honestly, the suction power is incredible for a cordless model, and it picked up pet hair from my thick rugs that my old upright vacuum completely missed. It's also super lightweight and easy to maneuver under the bed. However, the battery life is a massive letdown; it only lasts about 15 minutes on 'Max Power,' which isn't enough time to clean my whole apartment. Plus, the dustbin is way too small and requires emptying multiple times per cleaning session. It's a great machine if you have a small space, but frustrating for larger homes."
)
```

```python
from typing import Annotated, Optional, List

class Review2(TypedDict):
    review: Annotated[str, "The original raw text of the user's review."]
    sentiment: Annotated[str, "Must be one of: Positive, Negative, or Neutral."]
    key_points: Annotated[Optional[List[str]], "Bullet points summarizing the core feedback."]
```

```python
structured_llm2 = llm.with_structured_output(Review2)
```

```python
structured_llm2.invoke(
    "I’ve been using the AeroGlide Cordless Vacuum for about two weeks now. Honestly, the suction power is incredible for a cordless model, and it picked up pet hair from my thick rugs that my old upright vacuum completely missed. It's also super lightweight and easy to maneuver under the bed. However, the battery life is a massive letdown; it only lasts about 15 minutes on 'Max Power,' which isn't enough time to clean my whole apartment. Plus, the dustbin is way too small and requires emptying multiple times per cleaning session. It's a great machine if you have a small space, but frustrating for larger homes."
)
```

```python
from pydantic import BaseModel

class Review_pydantic(BaseModel):
  review: str
  sentiment: str
  pros: Optional[str]
  cons: Optional[str]
```

```python
structured_llm3 = llm.with_structured_output(Review_pydantic)
```

```python
structured_llm3.invoke(
    "I’ve been using the AeroGlide Cordless Vacuum for about two weeks now. Honestly, the suction power is incredible for a cordless model, and it picked up pet hair from my thick rugs that my old upright vacuum completely missed. It's also super lightweight and easy to maneuver under the bed. However, the battery life is a massive letdown; it only lasts about 15 minutes on 'Max Power,' which isn't enough time to clean my whole apartment. Plus, the dustbin is way too small and requires emptying multiple times per cleaning session. It's a great machine if you have a small space, but frustrating for larger homes."
)
```

```python
from pydantic import BaseModel, Field

class Review_pydantic2(BaseModel):
  review: str = Field(description = "The summary of original raw text of the user's review. ")
  product_name: str = Field(description = "The name of the product being reviewed. ")
  brand: Optional[str] = Field(description = "The brand of the product being reviewed. ")
  sentiment: str = Field(description = "Must be one of: Positive, Negative, or Neutral.")
  pros: Optional[str] = Field(description = "Bullet points summarizing the core positive feedback.")
  cons: Optional[str] = Field(description = "Bullet points summarizing the core negative feedback.")
```

```python
structured_llm4 = llm.with_structured_output(Review_pydantic2)
```

```python
ambrane_mouse_review = """
Ambrane offers a highly popular lineup of budget-friendly computer peripherals in India, focusing primarily on office productivity, student use, and casual setups. Most models cost under ₹500 while providing excellent everyday value.

## Overall Assessment of Ambrane Mice

Ambrane mice provide a high-value, no-frills experience based on typical user feedback and design specifications.

### What Users Like (Pros)
* **Silent Clicks:** Across almost all models, users frequently praise how quiet the left and right buttons are, making them ideal for shared offices, libraries, or late-night study sessions.
* **Affordable Pricing:** They are highly budget-friendly, making them easy to purchase as a secondary travel mouse or a quick, inexpensive replacement.
* **Portability & Design:** Most versions feature a very slim, lightweight profile that slides easily into a standard laptop bag sleeve.
* **Good Everyday Tracking:** The optical sensors provide smooth navigation on standard desks and basic mousepads for web browsing and office apps.

### User Complaints (Cons)
* **Build Quality Feel:** Some users note that the lightweight plastic shell can feel thin or hollow compared to higher-end brands.
* **Not for Serious Gaming:** The entry-level optical sensors suffer from slight latency or tracking jumps if you try to use them for fast-paced competitive gaming.
* **Durability:** Heavy everyday users mention that components like the scroll wheel or click tactile feel can sometimes degrade after several months of rough handling.

---

## Top Ambrane Mouse Models

Depending on your specific setup needs, Ambrane's wireless and wired lineup is divided into three primary categories:

### 1. Basic Productivity: Ambrane SliQ / SliQ 2
This is their most popular, entry-level wireless product. It is a traditional wireless mouse requiring an external AA or AAA battery.
* **Key Features:** It features silent click buttons, a standard 2.4GHz plug-and-play USB nano receiver, and up to 1600 adjustable DPI for tailored pointer speed.
* **Best For:** Laptop users who need a lightweight, quiet mouse to carry to cafes or meetings.

### 2. Premium Features: Ambrane SliQ Glow
A step up in terms of aesthetic flair and convenience, this model cuts out disposable batteries entirely.
* **Key Features:** It includes a Type-C rechargeable battery (lasting up to 15 days on a single charge), dual connectivity via Bluetooth 5.1 and 2.4GHz wireless, and vibrant RGB edge lighting.
* **Best For:** Users looking to sync a single mouse across multiple devices (like a laptop and a tablet) without swapping USB dongles.

### 3. Budget Wired Gaming: Ambrane TroneX
If you dislike managing wireless connections or charging batteries, this is a distinct wired option focused on casual gaming aesthetics.
* **Key Features:** It features an eye-catching honeycomb cutout design, a braided copper cable for durability, and 6 customizable buttons.
* **Best For:** Users wanting an inexpensive, wired mouse with an aggressive gaming style.

---

## Verdict: Should You Buy One?
* **Buy it if:** You are on a tight budget, need something highly portable, and deeply value completely silent click buttons for general office work or studying.
* **Skip it if:** You are a competitive gamer who requires flawless sensor precision, have very large hands that need a bulky ergonomic grip, or want premium, long-lasting materials.
"""
result = structured_llm4.invoke(ambrane_mouse_review)
```

```python
json_format = {
    "review": "The summary of original raw text of the user's review. ",
    "product_name": "The name of the product being reviewed. ",
    "brand": "The brand of the product being reviewed. ",
    "sentiment": "Must be one of: Positive, Negative, or Neutral.",
    "pros":{
        "type": "array",
        "description": "Bullet points summarizing the core positive feedback.",
        "properties": {
            "pro": {
                "type": "string",
                "description": "Bullet points summarizing the core positive feedback."
            }
        }
    },
    "cons":{
        "type": "array",
        "description": "Bullet points summarizing the core negative feedback.",
        "properties": {
            "con": {
                "type": "string",
                "description": "Bullet points summarizing the core positive feedback."
            }
        }
    },
}
```

```python
structured_llm5 = llm.with_structured_output(json_format)
```

```python
json_result = structured_llm5.invoke(ambrane_mouse_review)
```
